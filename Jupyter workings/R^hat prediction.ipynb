### Cleanup of above lines to generate prediction for a single condo

```python
def predict_condo(condo:str, n:int, df:pd.DataFrame, coeff_df:pd.DataFrame, metrics_df: pd.DataFrame):
    """
    Input: 
    condo: name of condo
    n: number of quarters to predict forward into
    df: dataframe of historical data
    coeff_df: dataframe containing coefficients of the condos
    metrics_df: dataframe containing metrics of regression of the condos
    
    
    Output:
    append to predict_dict
    """


    if condo == "District 15":
        condo_df = df.groupby("Project Name").last()
        year_prev = 2022  # update in the future
        q_prev = 4   # update in the future
        
        # starting changing variables
        age_q_prev = condo_df["Age_Q"].mean()  # assume as mean
        MRT = 1   # assume near school and mrt
        schools = 1   
        
        psf_latest = condo_df["Avg_PSF"].mean()  # assume as mean
        psf_prev = psf_latest
        
    else:
        condo_df = df[df["Project Name"]==condo]
        year_prev = condo_df["Year"].iloc[-1]
        q_prev = condo_df["Q"].iloc[-1]
        
        # starting changing variables
        age_q_prev = condo_df["Age_Q"].iloc[-1]
        MRT = condo_df["MRT"].iloc[-1]
        schools = condo_df["Schools"].iloc[-1]
        
        psf_latest = condo_df["Avg_PSF"].iloc[-1]
        psf_prev = psf_latest

    
    # starting fixed variables
    mean_dppi_percent = condo_df["DPPI_percent"].mean()
    mean_vol = condo_df["Vol"].mean()
    covid = 0    

    condo_metrics_df = metrics_df[metrics_df.iloc[:, 0]==condo]
    RMSE_testing = float(condo_metrics_df["RMSE (testing)"])
    RMSE_training = float(condo_metrics_df["RMSE (training)"])
    quarters_with_records = condo_df.count()[0]
    
    # fixed coeffs
    condo_coeff = np.array(coeff_df.loc[condo])

    # predict n number of quarters into the future
    for i in range(n):
        # update changing variables
        age_q_prev += 1
        if q_prev < 4:
            q_prev += 1
        else:
            q_prev = 1
            year_prev += 1

        # changing var_array
        var_array = np.array([1, 
                      mean_dppi_percent, 
                      math.log(age_q_prev),
                      MRT,
                      mean_dppi_percent*math.log(age_q_prev),
                      mean_dppi_percent*MRT,
                      math.log(mean_vol), 
                      mean_dppi_percent*math.log(mean_vol), 
                      covid,
                      mean_dppi_percent*covid,
                      math.log(mean_vol)*math.log(age_q_prev),
                      math.log(age_q_prev)*math.log(mean_vol)
                     ])
        delta_psf_pred = np.matmul(var_array.T, condo_coeff)
        
        psf_pred = psf_prev*(1 + delta_psf_pred)
        UB = psf_prev*(1 + delta_psf_pred + 1.96*RMSE_training)
        LB = psf_prev*(1 + delta_psf_pred - 1.96*RMSE_training)
        
        profit = (psf_pred - psf_latest)/psf_latest
        risk = ((UB-LB)*RMSE_testing)/quarters_with_records
        property_score = profit/risk
        
        # update psf_prev
        psf_prev = psf_pred

        # update prediction dictionary
        predict_dict["Project Name"].append(condo)
        predict_dict["delta_psf_pred"].append(delta_psf_pred)
        predict_dict["PSF_pred"].append(psf_pred)
        predict_dict["UB"].append(UB)
        predict_dict["LB"].append(LB)
        predict_dict["Profit"].append(profit)
        predict_dict["Risk"].append(risk)
        predict_dict["Property Score"].append(property_score)
        
        predict_dict["Year"].append(year_prev)
        predict_dict["Q"].append(q_prev)
        predict_dict["YYYYQQ"].append(f"{year_prev}Q{q_prev}")
        predict_dict["Mean DPPI_percent"].append(mean_dppi_percent)
        predict_dict["Mean Vol"].append(mean_vol)
        predict_dict["Age_Q"].append(age_q_prev)
        predict_dict["Covid"].append(covid)
        predict_dict["MRT"].append(MRT)
        predict_dict["Schools"].append(schools)
```

```python
# cleaned function code to run prediction on a single condo

# dict to store results of prediction
predict_dict = {"Project Name": [], "delta_psf_pred":[], "PSF_pred":[], "UB":[], "LB":[], "Profit":[], "Risk":[], "Property Score":[],
                "Year":[], "Q":[], "YYYYQQ":[], 
                "Mean DPPI_percent":[], "Mean Vol":[], 
                "Age_Q":[], "Covid":[], "MRT":[], "Schools":[]}

df = pd.read_csv("../Raw Data/ModelDB grp by.csv")
coeff_df = pd.read_csv("../Results/coeff.csv", index_col=".id")
metrics_df = pd.read_csv("../Results/metrics.csv")

predict_condo("THE SHORE RESIDENCES", 40, df, coeff_df, metrics_df)

pd.DataFrame(predict_dict).head()
```
