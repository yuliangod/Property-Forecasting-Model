```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
import time
import re
import pandas as pd
```

```python
#xpaths
closetutpath = '//*[@id="my-text"]/img'

searchbarpath = '//*[@id="search-text"]'

mrtstnpath='//*[@id="mrt_0"]'
mrtdistpath='//*[@id="distance"]/div[3]/div[2]'
LLpath='//*[@id="start_marker"]/div[1]/div[2]'
```

```python
#start chrome
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
#options.add_argument('--headless')
path="../Dependencies/chromedriver.exe"
driver = webdriver.Chrome(executable_path=path, options=options)

#open website
driver.get("https://www.onemap.gov.sg/main/v2/")

element = driver.find_element(By.XPATH, closetutpath)
element.click()
```

```python
df=pd.read_csv("../Databases/Scraped data/condos_attr.csv")
condonames=df["Project Name"].unique()

columns = ["Project Name","MRT", "MRT dist", "Lat", "Long"]
tempMRTdf=pd.DataFrame(columns=columns)

for condoname in condonames:
    searchbar = driver.find_element(By.XPATH, searchbarpath)
    searchbar.send_keys(condoname)
    searchbar.send_keys(Keys.RETURN)
    time.sleep(3)
    
    LLnew=driver.find_element(By.XPATH, LLpath).text.split(", ")
    if LLnew!=LL:
        LL=LLnew
        mrtstn=driver.find_element(By.XPATH, mrtstnpath).text
        mrtdist=driver.find_element(By.XPATH, mrtdistpath).text
        LL=driver.find_element(By.XPATH, LLpath).text.split(", ")
        tempMRTdf.loc[len(tempMRTdf)]=[condoname,mrtstn,mrtdist,LL[0],LL[1]]
        print (condoname,mrtstn,mrtdist,LL)
    else:
        print (condoname, " failed")
    searchbar.clear()
```
