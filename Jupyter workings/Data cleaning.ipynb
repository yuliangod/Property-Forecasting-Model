# Script to get average quarterly data to match with PPI

```python
import pandas as pd
```

## Clean property data

```python
sheet_name_ls = pd.ExcelFile("../Raw Data/condominium_edgeprop.xlsx").sheet_names
sheet_name_ls
```

```python
sheet_name = "PEBBLE BAY"
```

```python
prop_df = pd.read_excel("../Raw Data/condominium_edgeprop.xlsx", sheet_name=sheet_name)
```

```python
prop_df.head()
```

```python
type(prop_df["Date of Sale"][1])
```

Extracting Year and Month of each entry

```python
prop_df["Year"] = prop_df["Date of Sale"].dt.year
prop_df["Month"] = prop_df["Date of Sale"].dt.month
prop_df.head()
#type(prop_df["Month"][1])
```

Add column with which quarter entry is in

```python
#Q_dict = {"Jan":"Q1", "Feb": "Q1", "Mar": "Q1", "Apr": "Q2", "May": "Q2", "Jun": "Q2", "Jul":"Q3", "Aug":"Q3", "Sep":"Q3", "Oct":"Q4", "Nov":"Q4", "Dec":"Q4"}
Q_dict = {1:"Q1", 2: "Q1", 3: "Q1", 4: "Q2", 5: "Q2", 6: "Q2", 7:"Q3", 8:"Q3", 9:"Q3", 10:"Q4", 11:"Q4", 12:"Q4"}
```

```python
prop_df["Quarter"] = prop_df["Month"].map(Q_dict) #+ "-" + df["Year"]
```

```python
prop_df.head()
```

Group by quarter to get average data and count of each row

```python
prop_df = prop_df.groupby(["Year","Quarter"])
```

```python
count_series = prop_df["Project Name"].count()
count_series.head()
count_series
```

```python
prop_df = prop_df.mean()
prop_df.head()
```

Change No. of Units to count 

```python
prop_df["No. of Transactions"] = count_series
prop_df.head()
```

```python
#prop_df.index
```

## Clean PPI data

```python
ppi_df = pd.read_csv("../Raw Data/private-residential-property-price-index-base-quarter-2009-q1-100-quarterly.csv")
ppi_df.head()
```

Extracting Year and Month of each entry

```python
#ppi_df["Year"] = ppi_df["quarter"].str[2:4]
#ppi_df["Full Year"] = pd.to_numeric(ppi_df["quarter"].str[:4])
ppi_df["Year"] = pd.to_numeric(ppi_df["quarter"].str[:4])
ppi_df["Quarter"] = ppi_df["quarter"].str[-2:]
ppi_df.head()
```

Filter out irrelevant Years

```python
#ppi_df = ppi_df[ppi_df["Full Year"] <= 2020]
#ppi_df = ppi_df[ppi_df["Full Year"] >= 1990]
ppi_df = ppi_df[ppi_df["Year"] <= 2020]
ppi_df = ppi_df[ppi_df["Year"] >= 1990]
ppi_df.head()
```

Filter out only Non-Landed properties

```python
ppi_df = ppi_df[ppi_df["property_type"]=="Non-Landed"]
ppi_df.head()
```

```python
ppi_df = ppi_df.set_index(["Year", "Quarter"])
ppi_df.head()
```

```python
ppi_df.index
```

Rename index to PPI

```python
ppi_df = ppi_df.rename(columns={"index": "PPI"})
ppi_df
```

## Combining dataframes

```python
result_df = pd.merge(prop_df, ppi_df, how="right", left_index=True, right_index=True)
result_df
```

```python
import numpy as np
result_df["No. of Transactions in previous quarter"] = result_df["No. of Transactions"].shift(periods=1).replace(np.nan, 0)
result_df.tail(15)
```

Change index to datetime format and set as index

```python
#Year = list(result_df["Full Year"])
result_df = result_df.reset_index()
Year = list(result_df["Year"])
Year = [int(i) for i in Year]
Quarter = list(result_df["Quarter"].str[-1])
Quarter = [int(i) for i in Quarter]
Year
```

```python
result_df["Index"] = pd.PeriodIndex(year=Year, quarter=Quarter).to_timestamp()
result_df.tail()
```

```python
#result_df["Year"] = result_df["Full Year"]
#result_df = result_df.drop("Full Year", axis=1)
result_df = result_df.set_index("Index")
result_df
```

Add column delta PSF and delta PPI for regression to account for market noise

delta absolute

```python
delta_psf_abs = result_df[["Unit Price ($psf)", "PPI"]].dropna()["Unit Price ($psf)"].diff()
delta_ppi_abs = result_df[["Unit Price ($psf)", "PPI"]].dropna()["PPI"].diff()
result_df["delta_PSF"] = delta_psf_abs
result_df["delta_PPI"] = delta_ppi_abs
```

```python
result_df[["Unit Price ($psf)", "PPI"]].dropna()["Unit Price ($psf)"].diff()
result_df[["Unit Price ($psf)", "PPI"]].dropna()["PPI"].diff()
```

```python
result_df
```

delta PSF percentage

```python
delta_psf_pct = result_df[["Unit Price ($psf)", "PPI"]].dropna()["Unit Price ($psf)"].diff()/result_df["Unit Price ($psf)"].dropna().shift(periods=1)
delta_ppi_pct = result_df[["Unit Price ($psf)", "PPI"]].dropna()["PPI"].diff()/result_df["PPI"].dropna().shift(periods=1)
result_df["delta_PSF"] = delta_psf_pct
result_df["delta_PPI"] = delta_ppi_pct
```
